## 30. Identify any outliers in the movie runtime data. How would you handle them?

```python
plt.boxplot(df['Runtime (Minutes)'])
plt.show()
```

```python
outliers = []
def detect_outliers_iqr(data):
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    q2 = np.percentile(data, 50)
    print("q1: ",q1,"\nq2: ",q2,"\nq3: ", q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    print("lwr_bound: ",lwr_bound, "\nupr_bound",upr_bound)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers# Driver code
sample_outliers = detect_outliers_iqr(df['Runtime (Minutes)'])
print("Outliers from IQR method: ", sample_outliers)
#  print(lwr_bound, upr_bound)
```

```python
def manage_outliers(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    IQR = q3 - q1
    lower_bound = q1 - (1.5 * IQR)
    upper_bound = q3 + (1.5 * IQR)
    
    # Clip outliers to the upper and lower bounds
    data_clipped = np.clip(data, lower_bound, upper_bound)
    
    return data_clipped

# Apply the outlier management to your dataset column
df['Runtime (Minutes)'] = manage_outliers(df['Runtime (Minutes)'])
```

Technique Used 
Cap or Clip Outliers:
Instead of removing outliers, you can cap or clip their values to the lower and upper bounds of the IQR. This way, you retain the information while minimizing the impact of extreme values on your analysis
